```python
import time
import random
import math

from lineplot import LinePlot

plot = LinePlot(width="50%", y_ticks=3, colors=['red', 'gold', 'purple'])

for i in range(100):
    time.sleep(0.1)
    plot.add(loss=0.1+1/(0.5*i+1), random=random.random(), **{"1/(i+1)": 1/(i+1)})
```

```python

plot2 = LinePlot(colors=['#f00', '#a00', '#0f0', '#0a0'])

for i in range(20):
    time.sleep(0.1)
    plot2.add(
        acc=1 - random.random() / (i + 2.),
        val_acc=1 - random.random() / (i + 0.5),
        loss=1. / (i + 2.),
        val_loss=1. / (i + 0.5)
    )
```
